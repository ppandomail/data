* Cargo datos del archivo abalone.txt

```r
datos <- read.table("abalone.txt", header = F, sep = ",")
```

* Muestro las primeras 6 filas

```r
head(datos)
```

* Pongo nombre a las columnas

```r
colnames(datos) <- c("Genero", "Long", "Diam", "Altura", "Peso", "Peso1", "Peso2", "Peso3", "N.anillos")
head(datos)
```

* Muestro estructura de los datos

```r
str(datos)
```

* Convierto a los datos genero como factor. Variables cualitativas declararlas como factor (se reconocen categorias)

```r
datos$Genero <- as.factor(datos$Genero)
```

* Muestro resumen

```r
summary(datos)
```

* Muestro gráfico de barras para Género

```r
attach(datos) # es como with
frecG <- table(Genero)
barplot(frecG, 
        main = "Gráfico de barras para Género", 
        xlab = "Género", 
        ylab = "Frecuencias", 
        col = c("pink", "green", "blue"),
        legend = rownames(frecG))
```

* Muestro gráfico de torta o sectores

```r
pie (frecG, col = hcl.colors(length(frecG), "Fall"))
```

* Muestro frecuencias relativas

```r
prop.table(frecG)
```

* Muestro histograma y datos del histograma

```r
histo <- hist(Long, breaks = 12, 
        main = "Histograma de Longitud", 
        col = "lightblue", 
        border = "purple")
```

* Muestro boxplot

```r
boxplot(Long, 
        las = 1,
        col = "darkorchid1",
        ylab = "longitud")
```

```r
datos_num <- select_if(datos[,-1], is.numeric)
summary(datos_num)
```

```r
par(mfrow = c(2, 2))
boxplot(Long,
        col = "darkorchid1",
        xlab = "Longitud")
boxplot(Diam,
        col = "darkorchid1",
        xlab = "Diámetro")
boxplot(Altura,
        col = "darkorchid1",
        xlab = "Altura")
boxplot(Peso,
        col = "darkorchid1",
        xlab = "Peso")
```
