# Imputar datos

* Es completar los datos faltantes

```r
library(VIM)
library(mice)
```

```r
base <- airquality
head(base)
```

```r
aggr(base, numbers=T, sortVar=T)
```

```r
columnas <- c("Ozone", "Solar.R")
```

```r
imput_data <- mice(base[,names(base) %in% columnas],m = 1, maxit = 1, method = "mean", print=F)
```

```r
complete.data <- complete(imput_data)
```

```r
hist(na.omit(airquality$Ozone),
     breaks = 15,
     col = NA,
     border = "blue",
     ylim = c(0, 40),
     xlab = "Ozone (ppb)",
     ylab = "Frequency",
     main = "")
```

```r
hist(na.omit(airquality$Solar.R),
     breaks = 15,
     col = NA,
     border = "blue",
     ylim = c(0, 40),
     xlab = "Solar.R",
     ylab = "Frequency",
     main = "")
```
