```r
# importar la base de datos
df  <- read.csv2("properati (USD).csv", stringsAsFactors = TRUE)
head(df)
```

```r
# seleccionar variables (cols)
df1 <- df %>% select(place_name, price, year)
head(df1)
```

```r
# transformar la variable precios en dólares a pesos y agregarla a la tabla
df2 <- df1 %>% mutate(price_pesos = price*1000)
head(df2)
```

```r
# sort: ordena (default de menor a mayor)
niveles_barrios <- sort(unique(df$place_name))[1:10]
niveles_barrios
```

```r
# filtrar según condiciones
df3 <- df2 %>% filter(place_name %in% niveles_barrios & year > 2015)
head(df3)
```

```r
# ordenar en forma descendente el df a partir de la columna
df4 <- df3 %>% arrange(desc(price_pesos))
head(df4)
```

```r
# calcular promedio, minimo y máximo del precio en pesos por barrios
df5 <- df4 %>% group_by(place_name) %>%                               
                summarise(promedio_precio_pesos = mean(price_pesos), 
                min_precio_pesos = min(price_pesos), 
                max_precio_pesos = max(price_pesos))
head(df5)
```

```r
ancho <- df4 %>% group_by(place_name, year) %>% 
                summarise(precio_promedio = mean(price)) %>% 
                pivot_wider(names_from = year,values_from = precio_promedio)
head(ancho)
```

```r
library(gapminder)
```

¿Cuántos países hay en la base?

```r
gapminder <- gapminder::gapminder
View(gapminder)
```

```r
# alternativa 1
x <- length(unique(gapminder$country))
x
```

```r
# alternativa 2
x <- gapminder %>% 
        summarise(num_paises = n_distinct(country)) 
x
```

Indicar cuántos países hay por continente

```r
x <- gapminder %>% 
        group_by(continent) %>% 
        summarise(num_paises = n_distinct(country))  
x
```

¿Cuál es la observación con mayor expectativa al nacer de toda la base? ¿A qué país corresponde y a qué año?

```r
x <- gapminder %>% 
        filter(lifeExp == max(lifeExp)) %>% 
        select(country, year, lifeExp)
x
```

¿Cuál es el país más pobre y más rico, medido en PBI percápita, de cada continente para cada uno de los años?

```r
x <- gapminder %>% 
        group_by(continent, year) %>% 
        summarise(
          pais_mas_rico = country[which.max(gdpPercap)],
          gdp_mas_rico = max(gdpPercap, na.rm = TRUE),
          pais_mas_pobre = country[which.min(gdpPercap)],
          gdp_mas_pobre = min(gdpPercap, na.rm = TRUE),
          .groups = 'drop')
x
```

¿Cuál es la expectativa de vida a nacer promedio por continente en 1952? ¿Y en 2007?

```r
x <- gapminder %>% 
        filter(year %in% c(1952, 2007)) %>%   # Filtrar solo los años 1952 y 2007 
        group_by(continent, year) %>%         # Agrupar por continente y año
        summarise(expectativa_promedio = mean(lifeExp, na.rm = TRUE), .groups = 'drop') %>% 
        pivot_wider(names_from = year, values_from = expectativa_promedio, names_prefix = "expectativa_")
x
```

Realizar una tabla en formato "wide" en donde las columnas contengan los años, 
y las observaciones sean los promedios de la población de los paises de América

```r
x <- gapminder %>% 
        filter(continent == 'Americas') %>% 
        group_by(year) %>% 
        summarise(promedio_poblacion = mean(pop)) %>% 
        pivot_wider(names_from = year, values_from = promedio_poblacion)
x
```

¿Cuánto aumentó la expectativa de vida al nacer por continente entre 1952 y 2007?

```r
x <- gapminder %>% 
        filter(year %in% c(1952, 2007)) %>%  
        group_by(continent, year) %>%  
        summarise(exp_vida = mean(lifeExp), .groups = 'drop') %>%  
        pivot_wider(names_from = year, values_from = exp_vida) %>% 
        mutate(aumento = 2007 - 1952)
x
```

Presenta el PIB pércapita de Argentina para cada año menos para los años 1952 y 2007

```r
x <- gapminder %>% 
        select(country, year, gdpPercap) %>% 
        filter(country == "Argentina") %>% 
        filter(!(year %in% c(1952, 2007)))
x
```

Obtener para cada país la máxima expectativa de vida de los diferentes años, luego ordenarlos de menor a mayor

```r
x <- gapminder %>% 
        select(country, year, lifeExp) %>% 
        group_by(country) %>% 
        summarise(max_exp = max(lifeExp)) %>% 
        arrange(max_exp)
x
```
