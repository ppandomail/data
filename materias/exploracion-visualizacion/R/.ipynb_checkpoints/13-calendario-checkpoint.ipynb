```r
library(calendR)
library(readxl)
library(tidyverse)
library(dplyr)
library(tidyr)
```

```r
setwd("/Users/ppando/Materias/data/materias/exploracion-visualizacion/R")
```

```r
covid <- read_excel("fallecidos de covid.xlsx")
head(covid)
```

```r
covid_2021 <- covid %>% filter(dia > "2020-12-31")
head(covid_2021)
```

```r
calendR(year = 2021,
        special.days = covid_2021$total,
        gradient = TRUE,
        low.col = "white",
        special.col = "#FF4600",
        legend.pos = "right",
        legend.title = "Cantida de fallecidos") 
```

```r
library(googleVis)
```

```r
plotdata <- gvisCalendar(
    data=covid, datevar="dia", numvar="total", 
    options=list(title="Fallecimientos por Coronavirus en Argentina", 
                 calendar="{cellSize:10, yearLabel:{fontSize:20, color:'#444444'}, focusedCellColor:{stroke:'red'}}",
                 width=700, 
                 height=400),
    chartid="Calendar")

plot(plotdata)
```
